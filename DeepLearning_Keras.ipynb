```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
import numpy as np
import tensorflow as tf
```

```python
seed = 777
np.random.seed(seed)
```

```python
tf.set_random_seed(seed)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/DeepLearning_Data/Seoul_Vacancy_Rate_v4.csv', encoding= 'ansi')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_data = df.iloc[:, :-1]
x_data.head()
```

```python
y_data = df.iloc[:, -1:]
y_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size= 0.3,
                                                   random_state= seed)
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(24, input_dim= 42, activation= 'relu'))
```

```python
model.add(Dense(10, activation= 'relu'))
```

```python
model.add(Dense(10, activation= 'relu'))
```

```python
model.add(Dense(1))
```

```python
model.compile(loss= 'mean_squared_error', optimizer= 'adam')
```

```python
model.fit(X_train, y_train, epochs= 50, batch_size= 5)
```

```python
Y_predicion = model.predict(X_test).flatten()
```

```python
a = y_test.values
a[1,0]
```

```python
for i in range(20):
    label = a[i, 0]
    prediction = Y_predicion[i]
    print('{:.3f}, ==== {:.3f}'.format(label, prediction))
```
